```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt

print("loading dataset...")

labeled_dataset = pd.read_csv('data/labeled_processed.csv')

print("dataset loaded!")

fields_to_drop = [
    'sample', 'file_name', 'ML_1', 'ML_1_1', 'ML_1_1_1', 'ML_1_2','ML_2','ML_2_1_1','ML_2_1_2','ML_2_1_3','ML_2_2','ML_2_3',
    'ML_2_3_1','ML_2_4_1','ML_2_4_2','ML_2_5_1','ML_2_6','ML_2_7_1','ML_2_7_2','ML_3','ML_3_1','ML_3_2','ML_3_3','ML_3_4',
    'ML_3_5', 'event_type']

all_features = labeled_dataset.drop(fields_to_drop, axis=1)
all_labels = labeled_dataset['event_type']
```

```python
classifier = CatBoostClassifier()
classifier.load_model("trained_osc_grid_catboost")

train_dataset_score = classifier.score(all_features, all_labels)
print(f"Score Accuracy: {train_dataset_score}")

```
